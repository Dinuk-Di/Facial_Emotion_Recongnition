```python
import webbrowser
import psutil
import win32gui
import win32con
import win32process
import time

def is_browser_running(browser_name):
    for proc in psutil.process_iter(['pid', 'name']):
        if proc.info['name'] and browser_name.lower() in proc.info['name'].lower():
            return proc.info['pid']
    return None

def bring_browser_to_front(pid):
    def enumHandler(hwnd, lParam):
        if win32gui.IsWindowVisible(hwnd):
            _, found_pid = win32process.GetWindowThreadProcessId(hwnd)
            if found_pid == pid:
                win32gui.ShowWindow(hwnd, win32con.SW_RESTORE)
                win32gui.SetForegroundWindow(hwnd)
                return False
        return True
    win32gui.EnumWindows(enumHandler, None)

def open_url_in_browser(url, browser_path=None):
    if browser_path:
        webbrowser.register('custom', None, webbrowser.BackgroundBrowser(browser_path))
        browser = webbrowser.get('custom')
    else:
        browser = webbrowser
    browser.open(url, new=2)

url = "https://youtube.com"
browser_exe = "chrome.exe"
chrome_path = r"C:\Program Files\Google\Chrome\Application\chrome.exe"

pid = is_browser_running(browser_exe)
open_url_in_browser(url, browser_path=chrome_path)

# Sleep to allow browser to open new tab
time.sleep(2)

if pid:
    bring_browser_to_front(pid)
else:
    print("Browser was not running previously.")
```

```python
import sys, subprocess, time, urllib.request, json
import psutil, win32gui, win32con, win32process
from websocket import create_connection
import webbrowser, random, re

def start_chrome_debug(chrome_exe, data_dir=None):
    # Launch an entirely new Chrome profile with remote debugging.
    port = random.choice([9222, 9333, 9444])
    cmd = [chrome_exe, f"--remote-debugging-port={port}"]
    if data_dir:
        cmd.append(f"--user-data-dir={data_dir}")
    subprocess.Popen(cmd)
    time.sleep(3)
    return port

def open_new_tab_via_cdp(port, url):
    info = urllib.request.urlopen(f"http://localhost:{port}/json/new?{url}", timeout=3)
    tab = json.loads(info.read().decode())
    return tab["id"], tab["webSocketDebuggerUrl"]

def close_target_via_cdp(ws_url, target_id):
    ws = create_connection(ws_url)
    ws.send(json.dumps({
        "id": 1,
        "method": "Target.closeTarget",
        "params": {"targetId": target_id}
    }))
    ws.close()
    return True

def is_chrome_running(exe_name="chrome.exe"):
    for p in psutil.process_iter(['pid', 'name']):
        if p.info['name'] and exe_name.lower() in p.info['name'].lower():
            return p.info['pid']
    return None

def bring_to_front(pid):
    def handler(hwnd, _):
        if win32gui.IsWindowVisible(hwnd):
            _, found = win32process.GetWindowThreadProcessId(hwnd)
            if found == pid:
                win32gui.ShowWindow(hwnd, win32con.SW_RESTORE)
                win32gui.SetForegroundWindow(hwnd)
                return False
        return True
    win32gui.EnumWindows(handler, None)

def main():
    chrome_exe = r"C:\Program Files\Google\Chrome\Application\chrome.exe"
    custom_profile = r"C:\temp\cdp-profile"
    url = "https://youtube.com"
    
    pid = is_chrome_running()
    debug_port = None
    
    # Try existing debug or launch new Chrome
    for p in [9222, 9333, 9444]:
        try:
            tabs = urllib.request.urlopen(f"http://localhost:{p}/json", timeout=1)
            debug_port = p
            break
        except:
            continue
    if not debug_port:
        debug_port = start_chrome_debug(chrome_exe, data_dir=custom_profile)
    
    print(f"[+] Using CDP port {debug_port}")
    
    target_id, ws_url = open_new_tab_via_cdp(debug_port, url)
    print(f"[+] YouTube opened in tab ID: {target_id}")
    
    print("[*] Waiting 5 seconds before closing YouTube tab...")
    time.sleep(5)
    
    if close_target_via_cdp(ws_url, target_id):
        print("[✓] YouTube tab closed cleanly.")
    else:
        print("[!] Failed to close the tab via CDP.")
    
    if pid:
        time.sleep(1)
        bring_to_front(pid)

if __name__ == "__main__":
    main()
```

```python
import subprocess

def is_focus_assist_on():
    # This command queries the registry key for Focus Assist (Quiet Hours) state
    cmd = [
        "powershell",
        "-Command",
        "Get-ItemPropertyValue -Path 'HKCU:\\Software\\Microsoft\\Windows\\CurrentVersion\\Notifications\\Settings' -Name 'NOC_GLOBAL_SETTING_TOASTS_ENABLED'"
    ]
    result = subprocess.run(cmd, capture_output=True, text=True)
    # Registry value 0 means focus assist/do not disturb enabled, 1 means disabled
    if result.returncode == 0:
        state = result.stdout.strip()
        return state == "0"
    return False

if __name__ == "__main__":
    if is_focus_assist_on():
        print("Do Not Disturb (Focus Assist) is ON")
    else:
        print("Do Not Disturb (Focus Assist) is OFF")
```

```python
import win32gui
import win32process
import psutil
import os
import signal

def is_window_visible_and_has_title(hwnd):
    return win32gui.IsWindowVisible(hwnd) and win32gui.GetWindowText(hwnd)

def get_open_apps_with_pids():
    apps = []

    def enum_window_callback(hwnd, _):
        if is_window_visible_and_has_title(hwnd):
            window_title = win32gui.GetWindowText(hwnd)
            _, pid = win32process.GetWindowThreadProcessId(hwnd)
            apps.append((window_title, pid))

    win32gui.EnumWindows(enum_window_callback, None)
    return apps

def close_app_by_pid(pid):
    try:
        # Using psutil to terminate the process more gracefully
        proc = psutil.Process(pid)
        proc.terminate()  # sends SIGTERM equivalent
        proc.wait(timeout=5)  # wait for it to exit
        print(f"Process {pid} terminated.")
    except psutil.NoSuchProcess:
        print(f"No process found with PID {pid}.")
    except psutil.TimeoutExpired:
        print(f"Process {pid} did not terminate in time, killing it.")
        proc.kill()  # force kill
    except Exception as e:
        print(f"Error terminating process {pid}: {e}")

if __name__ == "__main__":
    apps = get_open_apps_with_pids()
    print("Open apps with PID:")
    for title, pid in apps:
        print(f"Title: {title}, PID: {pid}")
```

```python
import win32gui
import win32process
import psutil

def get_chrome_tabs():
    chrome_tabs = []

    def enum_windows(hwnd, _):
        # Check if window is visible and has a title
        if win32gui.IsWindowVisible(hwnd):
            title = win32gui.GetWindowText(hwnd)
            if "Chrome" in title or "Google Chrome" in title:
                _, pid = win32process.GetWindowThreadProcessId(hwnd)
                # Collect tab info: (Window Title, PID)
                chrome_tabs.append((title, pid))

    win32gui.EnumWindows(enum_windows, None)
    # Remove duplicates by PID (some tabs or windows may have same title)
    unique_tabs = {}
    for title, pid in chrome_tabs:
        unique_tabs[pid] = title
    return [(title, pid) for pid, title in unique_tabs.items()]

def kill_chrome_tab_by_pid(pid):
    try:
        proc = psutil.Process(pid)
        # Terminate the tab process (tab close)
        proc.terminate()
        proc.wait(timeout=5)
        print(f"Chrome tab with PID {pid} closed.")
    except psutil.NoSuchProcess:
        print(f"No process found with PID {pid}. It might already be closed.")
    except psutil.TimeoutExpired:
        print(f"Process {pid} did not close in time, killing it.")
        proc.kill()
    except Exception as e:
        print(f"Error while closing tab PID {pid}: {e}")

if __name__ == "__main__":
    tabs = get_chrome_tabs()
    print("Open Chrome tabs found:")
    for title, pid in tabs:
        print(f"Title: {title}, PID: {pid}")

    # Example of killing a tab by PID (uncomment carefully)
    # if tabs:
    #     kill_chrome_tab_by_pid(tabs[0][1])
```

```python
import asyncio
from pycdp import cdp
from pycdp.browser import ChromeLauncher
from pycdp.asyncio import connect_cdp

async def main():
    chrome = ChromeLauncher(
        binary=r'C:\\Program Files\\Google\\Chrome\\Application\\chrome.exe',
        args=['--remote-debugging-port=9222']
    )
    await asyncio.get_running_loop().run_in_executor(None, chrome.launch)
    conn = await connect_cdp('http://localhost:9222')
    
    # Create a new tab or get existing tab via target.create_target
    target_id = await conn.execute(cdp.target.create_target('about:blank'))
    session = await conn.connect_session(target_id)
    await session.execute(cdp.page.enable())
    
    # To close: 
    await conn.execute(cdp.target.close_target(target_id))

asyncio.run(main())
```

```python
import os
import subprocess
import psutil
import win32com.shell.shell as shell
import win32con
import tkinter.messagebox as messagebox

def find_executable_in_folder(folder_path):
    if not os.path.isdir(folder_path):
        return None

    exe_files = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(".exe"):
                exe_files.append(os.path.join(root, file))

    if exe_files:
        exe_files.sort(key=lambda x: len(x))
        return exe_files[0]

    return None

def get_classic_installed_programs():
    """
    Returns a list of tuples: (DisplayName, InstallLocation)
    """
    import winreg

    uninstall_keys = [
        r"SOFTWARE\Microsoft\Windows\CurrentVersion\Uninstall",
        r"SOFTWARE\WOW6432Node\Microsoft\Windows\CurrentVersion\Uninstall"
    ]
    results = []

    for root in (winreg.HKEY_LOCAL_MACHINE, winreg.HKEY_CURRENT_USER):
        for subkey in uninstall_keys:
            try:
                with winreg.OpenKey(root, subkey) as key:
                    for i in range(0, winreg.QueryInfoKey(key)[0]):
                        try:
                            subkey_name = winreg.EnumKey(key, i)
                            with winreg.OpenKey(key, subkey_name) as app_key:
                                name = winreg.QueryValueEx(app_key, "DisplayName")[0]
                                location = winreg.QueryValueEx(app_key, "InstallLocation")[0] if "InstallLocation" in [winreg.EnumValue(app_key, j)[0] for j in range(winreg.QueryInfoKey(app_key)[1])] else ""
                                if name:
                                    results.append((name, location))
                        except:
                            continue
            except:
                continue
    return results

def get_installed_uwp_apps():
    """Returns a list of (AppName, AppUserModelID) using PowerShell"""
    try:
        command = [
            "powershell",
            "-NoProfile",
            "-ExecutionPolicy", "Bypass",
            "-Command",
            "Get-StartApps | ForEach-Object { \"$($_.Name)|$($_.AppID)\" }"
        ]
        output = subprocess.check_output(command, text=True)
        apps = []
        for line in output.strip().splitlines():
            if "|" in line:
                name, app_id = line.split("|", 1)
                apps.append((name.strip(), app_id.strip()))
        return apps
    except subprocess.CalledProcessError as e:
        print(f"[UWP Fetch Error] PowerShell returned error code {e.returncode}")
        print(f"Output: {e.output}")
        return []
    except Exception as e:
        print(f"[UWP Fetch Exception] {e}")
        return []

def search_installed_apps(name_input):
    name = name_input.lower()

    # Search UWP apps
    uwp_matches = [app for app in get_installed_uwp_apps() if name in app[0].lower()]
    if uwp_matches:
        print(f"[Match] Found UWP app: {uwp_matches[0][0]}")
        return {"type": "uwp", "name": uwp_matches[0][0], "app_id": uwp_matches[0][1]}

    # Search classic apps
    classic_matches = [app for app in get_classic_installed_programs() if name in app[0].lower()]
    if classic_matches:
        install_location = classic_matches[0][1]
        exe_path = find_executable_in_folder(install_location)
        return {"type": "classic", "name": classic_matches[0][0], "path": exe_path or install_location}

    return None
```

```python
def unified_app_launcher(app_info):
    try:
        if app_info["type"] == "uwp":
            app_id = app_info["app_id"]
            subprocess.Popen(f'explorer shell:AppsFolder\\{app_id}', shell=True)
            print(f"[Launch] UWP app: {app_info['name']}")
            return True

        elif app_info["type"] == "classic":
            exe_path = app_info["path"]
            if not exe_path or not os.path.isfile(exe_path):
                print(f"[Error] Executable not found: {exe_path}")
                return False
            
            shell.ShellExecuteEx(
                    lpVerb='runas',
                    lpFile=exe_path,
                    lpParameters='',
                    nShow=win32con.SW_SHOWNORMAL
            )
            print(f"[Launch] Classic app as Admin: {exe_path}")
            return True

    except Exception as e:
        print(f"[Launch Error] {e}")
        return False
```

```python
unified_app_launcher({'type': 'uwp', 'name': 'Discord', 'app_id': 'com.squirrel.Discord.Discord'})
```

```python
import win32gui
import win32process
import psutil
import os
import signal

def is_window_visible_and_has_title(hwnd):
    return win32gui.IsWindowVisible(hwnd) and win32gui.GetWindowText(hwnd)

def get_all_running_apps():
    apps = []
    for proc in psutil.process_iter(['pid', 'name', 'exe']):
        try:
            name = proc.info['name']
            pid = proc.info['pid']
            exe = proc.info['exe']
            apps.append((name, pid, exe))
        except (psutil.NoSuchProcess, psutil.AccessDenied):
            continue
    return apps

def close_app_by_pid(pid):
    try:
        # Using psutil to terminate the process more gracefully
        proc = psutil.Process(pid)
        proc.terminate()  # sends SIGTERM equivalent
        proc.wait(timeout=5)  # wait for it to exit
        print(f"Process {pid} terminated.")
    except psutil.NoSuchProcess:
        print(f"No process found with PID {pid}.")
    except psutil.TimeoutExpired:
        print(f"Process {pid} did not terminate in time, killing it.")
        proc.kill()  # force kill
    except Exception as e:
        print(f"Error terminating process {pid}: {e}")

if __name__ == "__main__":
    apps = get_all_running_apps()
    print("Open apps with PID:")
    for name, pid, exe in apps:
        print(f"Name: {name}, PID: {pid}, Executable: {exe}")
```
